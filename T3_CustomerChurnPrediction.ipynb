#**Logistic Regression, Random Forests, Gradient Boosting**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingClassifier

```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for column in data.columns:
    if data[column].dtype == 'object':
        data[column] = encoder.fit_transform(data[column])
```

```python
data.info()
```

```python
data.head()
```

```python
data = data.drop(['RowNumber', 'CustomerId', 'Surname'],axis=1)
data
```

```python
X = data.drop('Exited',axis=1)
y = data['Exited']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test)
```

```python
print('Accuracy-LogisticRegression')
y_test = [1 if value > threshold else 0 for value in y_test]

lr_score=lr.score(X_test,y_test)*100
lr_score
```

```python
threshold = 0.5
y_train_classified = [1 if value > threshold else 0 for value in y_train]
rf = RandomForestClassifier()
rf.fit(X_train, y_train_classified)
```

```python
y_test_classified = [1 if value > threshold else 0 for value in y_test]
rf_score = rf.score(X_test, y_test_classified)*100
rf_score
```

```python
threshold = 0.5
y_train_classified = [1 if value > threshold else 0 for value in y_train]
GBC = GradientBoostingClassifier()
GBC.fit(X_train, y_train_classified)
```

```python
y_test_classified = [1 if value > threshold else 0 for value in y_test]
GBC_score= GBC.score(X_test, y_test_classified)*100
GBC_score
```

```python
performance_summary = pd.DataFrame({
    'Model':['Logistic Regression','RandomForest','GradientBoosting'],
    'Accuracy':[lr_score,
           rf_score,
           GBC_score,
          ]
})
performance_summary
```
